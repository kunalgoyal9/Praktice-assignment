```python
data2 = []
for idx in range(len(data)):
    try:
        string = data[idx]
        s = string.split('\r\n')
        rmv = []
#         print(s)
        for i in range(len(s)-1):
            if s[i] != '' and s[i+1]!='':
                s[i] += ' ' + s[i+1]
                rmv.append(s[i+1])
                i+=1
#         print(rmv)
    #     s = list(s)
        for i in range(len(rmv)):
            if(len(s)>0):
                s.remove(rmv[i])
        data2.append(s)
#         print(data2)
    except:
        pass
```

```python
for i in range(len(data2)):
    data2[i] = data2[i].remove('')
#     print(data2[i].shape)
```
